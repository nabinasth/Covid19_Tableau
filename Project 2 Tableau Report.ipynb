```python
# drop unusable columns for data exploration
columns_to_drop = [
    'tests_units', 'new_tests_smoothed', 'new_tests_smoothed_per_thousand',
    'new_people_vaccinated_smoothed', 'new_people_vaccinated_smoothed_per_hundred',
    'excess_mortality_cumulative_absolute', 'excess_mortality_cumulative',
    'excess_mortality', 'excess_mortality_cumulative_per_million',
    'hospital_beds_per_thousand', 'female_smokers', 'male_smokers', 'handwashing_facilities',
    'new_deaths_smoothed_per_million', 'new_cases_smoothed_per_million','icu_patients_per_million',
    'hosp_patients_per_million','weekly_icu_admissions_per_million','weekly_hosp_admissions_per_million'
]
df.drop(columns= columns_to_drop,inplace= True)
```

```python
df.info()
```

```python
# Add a new column to analyze pre- and post-vaccine effects.
first_vaccine_date = df[df['total_vaccinations'].notna()]['date'].min()
first_vaccine_date
```

```python
df['vaccine_period'] = df['date'].apply(lambda x: 'Pre-vaccine' if x < first_vaccine_date else
                                        'Post-vaccine')
```

Exploratory Data Visualization and Dashboard Overview
Project Tool Used: Tableau

This dashboard explores trends and patterns in COVID-19 data across continents and years, focusing on cases, deaths, and vaccination efforts.
1. Continent Fully Vaccinated:
   This chart displays the number of people who were fully vaccinated by continent.
   Europe and North America had the highest number of fully vaccinated people.
   Filtered for the post-vaccine period only.





2. People Vaccinated per Year per Continent:
   This clustered bar chart compares yearly vaccination counts for each continent from 2020 to 2022.Asia had the largest vaccination peak in 2021.

 3. Cases, Deaths, and Vaccination per Continent:
    This line chart displays three metrics per continent: total cases, total deaths, and total vaccinations.Asia reported the highest vaccination and death counts.There is no direct proportionality between total cases and vaccinations across continents.



4. Total Cases and Deaths per Year:
   Yearly trend of total cases and deaths globally from 2020 to 2022.Both total cases and deaths peaked in 2021.
